```python
import os
# traverse whole directory

paths = []

for root, dirs, files in os.walk('../datalake/movie/'):
    # select file name
    for file in files:
        # check the extension of files
        if file.endswith('.json'):
            # print whole path of files
            print(os.path.join(root, file))
            paths.append(os.path.join(root, file))
```

```python
import json
```

```python
def write_to_json(path, list_data):
    with open(path, 'w') as file:
        json.dump(list_data, file, indent=2)
```

```python
def get_list_movie(path):
    # Opening JSON file
    try:
        f = open(path)
        data = json.load(f)
        print('Load data success in' + path)
        return data
    except:
        return None
```

```python
list_movie = []
for path in paths:
    list_data = get_list_movie(path)
    if list_data is not None:
        list_movie.extend(list_data)

write_to_json('./movie.json', list_movie)
```

## Feature engineering

```python
df.head()
```

```python
## Drop production countries 
df.drop(['Production Countries'], axis=1, inplace=True)
df.head()
```

```python
## Add profit column
df['profit'] = df['Revenue'] - df['Budget']
```

```python
df['ROI'] = (df['profit'] / df['Budget']) * 100
```

```python
df.head()
```

```python
#Make the numbers more readable by dividing budget, revenue and profits by a miliion
df['Budget'] = (df['Budget'] / 1000000).astype(int)
df['Revenue'] = (df['Revenue'] / 1000000).astype(int)
df['profit'] = (df['profit'] / 1000000).astype(int)

df.head()
```

```python
## Convert ROI into a differnt format in terms of (4.1x budget)
df['ROI'] = (df['ROI'] / 100).round(1)
df.head()
```

```python
## Rename and convert release_date into release_year, month and days
df['Year'] = df['Release Date'].dt.year
df['Month'] = df['Release Date'].dt.month
df['Day'] = df['Release Date'].dt.day
```

```python
df.head()
```

```python
df.columns
```

```python
##Rename columns 
new_cols = {'Title':'title','Overview':'overview','Release Date':'release_date', 'Vote Average':'vote_average','Vote Count':'vote_count','Runtime':'runtime','Budget':'budget','Revenue':'revenue','Popularity':'popularity',
            'Production Companies':'production_companies',
            'Genres':'genres',
            'Popularity':'popularity'}

df.rename(columns=new_cols, inplace=True)
df.head()
```

```python
## Rearrange the columns as needed
df = df[['title', 'overview', 'release_date','Year', 'Month','Day', 'vote_average', 'vote_count', 'popularity','production_companies', 'runtime', 'budget', 'revenue', 'profit', 'ROI',
        'genres']]

df.head()
```

## Perform one-hot-encoding on genres 

```python
genre_df = pd.DataFrame()
genre_df
```

```python
# List of major genres
major_genres = ['Action', 'Adventure', 'Science Fiction', 'Drama', 'Thriller', 'Comedy', 'Animation', 'Fantasy', 'Horror', 'Romance', 'Mystery', 'Documentary', 'Music', 'Western', 'Crime']
```

```python
# Iterate through major genres and populate the columns
for genre in major_genres:
    genre_df[genre] = df['genres'].str.contains(genre, case=False).astype(int)
```

```python
genre_df
```

```python
# Iterate over rows and check if they contain all zeros
found_all_zeros = False  # Initialize a flag to check if any rows contain all zeros

for index, row in genre_df.iterrows():
    if (row == 0).all():
        print(f"Row {index} contains all zeros.")
        found_all_zeros = True

if not found_all_zeros:
    print("There are no rows containing all zeros.")
```

```python
# Merge genre_df with df_main based on index
merged_df = df.join(genre_df)
merged_df
```

```python
#Example of using python to perform EDA (single genre)

# Filter the DataFrame for selected movie genre (Animation) and release year (2010)
selected_genre = merged_df[(merged_df['Action'] == 1) & (merged_df['Year'] == 2022)]

# Sort the filtered DataFrame by profit in descending order
top_10_movie = selected_genre.sort_values(by='profit', ascending=False)

# Select the top 10 thriller movies based on profit
top_10_movie = top_10_movie.head(10)

# Print the top 10 thriller movies
print(top_10_movie[['title', 'profit']])
```

```python
#Example of using python to perform EDA (multiple-genre)

# Filter the DataFrame for selected movie genre (Animation) and release year (2010)
selected_genre = merged_df[(merged_df['Action'] == 1) & (merged_df['Comedy'] == 1) & (merged_df['Year'] == 2022)]

# Sort the filtered DataFrame by profit in descending order
top_10_movie = selected_genre.sort_values(by='profit', ascending=False)

# Select the top 10 thriller movies based on profit
top_10_movie = top_10_movie.head(10)

# Print the top 10 thriller movies
print(top_10_movie[['title', 'profit']])
```
